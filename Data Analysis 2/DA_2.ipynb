## **Studying depression occurence and implications for different factors**

```python
# Plot histograms for 'Age' with hue as 'depression_category'
sns.histplot(data=df[df['depression_category'] == 'No Depression'], x='Age', color='blue',  kde=True)
sns.histplot(data=df[df['depression_category'] == 'Mild Depression'], x='Age', color='yellow', kde=True)
sns.histplot(data=df[df['depression_category'] == "Severe Depression"], x='Age', color='green',  kde=True)

# Set labels and title
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Age by Depression Category')
plt.legend(title='Depression Category',labels=['No Depression','Mild Depression','Severe Depression'])

# Show the plot
plt.show()
```

**Normalised Plot**

```python
# Compute the count of 'No Depression' category to normalise histograms
no_depression_count = df[df['depression_category'] == 'No Depression']['Age'].count()
mild_depression_count = df[df['depression_category'] == 'Mild Depression']['Age'].count()
severe_depression_count = df[df['depression_category'] == 'Severe Depression']['Age'].count()


# Plot histograms for 'Age' with hue as 'depression_category'
sns.histplot(data=df[df['depression_category'] == 'No Depression'], x='Age', color='blue',  kde=True, stat='density',weights=no_depression_count/no_depression_count)
sns.histplot(data=df[df['depression_category'] == 'Mild Depression'], x='Age', color='yellow', kde=True,stat='density',weights=mild_depression_count/no_depression_count)
sns.histplot(data=df[df['depression_category'] == "Severe Depression"], x='Age', color='green',  kde=True,stat='density',weights=severe_depression_count/no_depression_count)

# Set labels and title
plt.xlabel('Age')
plt.ylabel('Density')
plt.title('Distribution of Age by Depression Category')
plt.legend(title='Depression Category',labels=['No Depression','Mild Depression','Severe Depression'])

# Show the plot
plt.show()
```

**1. Age Group Concentration**: The most prominent age group affected by depression in category '0' (no depression) appears to be in the 40-60 age range, as indicated by the height of the blue bars. This suggests that a large portion of the population in this age group may not be experiencing depression, or at least not at levels categorized as '1' or '2'.

**2. Depression Severity Distribution:** The graph shows that as the severity of depression increases (from category '0' to '2'), the frequency of individuals in each category decreases. This could indicate that severe depression (category '2') is less common than no depression or mild/moderate depression.

**3. Age Trend for Severe Depression**: The yellow bars, which likely represent severe depression (category '2'), are not only lower in frequency but also show less variation with age compared to the other categories. This could imply that severe depression affects a relatively constant proportion of individuals across different ages, but it's also possible that the sample size for this category is smaller, which could affect the variability.

**4. Middle-Aged Focus**: There's a notable peak in the frequency of the blue bars (category '0') around the age of 55, suggesting that individuals in this age range are particularly likely to be classified in this category. Conversely, there is a noticeable dip in the frequency for the same category around the age of 65, which might indicate a transition point where individuals begin to experience more depression as they move into older age groups.

**5. Young Adults and Older Individuals**: The graph indicates that both the youngest (around 20) and oldest (around 80) age groups have the lowest frequencies of depression across all categories. However, it's worth considering that this could be due to lower population sizes in those age groups within the data set.

**6. Trend Line Analysis**: The blue trend line, which smooths out the variations across age groups, suggests that there's a general increase in the frequency of depression cases from around age 20 to a peak at about age 55, followed by a gradual decrease. However, there is a slight uptick again after age 70.

**8. Data Sparsity**: The heights of the bars for categories '1' and '2' at the extremes of the age range are much lower, which could indicate that there are fewer individuals in these age groups or that depression is less common there. The data seems to be more sparse for these categories in the youngest and oldest age groups.

**9. Clinical Implications**: From a public health perspective, the higher frequencies of depression in the middle-aged groups might indicate a target demographic for mental health services and interventions.

### **No of Hours of Sleep vs depression category**

```python
# Plot histograms for sleep with hue as 'depression_category'
sns.histplot(data=df[df['depression_category'] == 'No Depression'], x='No_of_hours_you_sleep', color='blue', label='No Depression', kde=True,)
sns.histplot(data=df[df['depression_category'] == 'Mild Depression'], x='No_of_hours_you_sleep', color='yellow', label='Mild Depression', kde=True)
sns.histplot(data=df[df['depression_category'] == 'Severe Depression'], x='No_of_hours_you_sleep', color='green', label='Severe Depression', kde=True)

# Set labels and title
plt.xlabel('No of hours slept')
plt.ylabel('Frequency')
plt.title('Distribution of Hours Slept by Depression Category')
plt.legend(title='Depression Category')

# Show the plot
plt.show()
```

```python
# Compute the count of 'No Depression' category to normalise histograms

no_depression_count = df[df['depression_category'] == 'No Depression']['No_of_hours_you_sleep'].count()
mild_depression_count = df[df['depression_category'] == 'Mild Depression']['No_of_hours_you_sleep'].count()
severe_depression_count = df[df['depression_category'] == 'Severe Depression']['No_of_hours_you_sleep'].count()

# Maximum count among the three classes
max_count = max(no_depression_count, mild_depression_count, severe_depression_count)

# Plot histograms for sleep with hue as 'depression_category'
sns.histplot(data=df[df['depression_category'] == 'No Depression'], x='No_of_hours_you_sleep', color='blue', label='No Depression', kde=True,stat='density',weights=no_depression_count/no_depression_count)
sns.histplot(data=df[df['depression_category'] == 'Mild Depression'], x='No_of_hours_you_sleep', color='yellow', label='Mild Depression', kde=True,stat='density',weights=mild_depression_count/no_depression_count)
sns.histplot(data=df[df['depression_category'] == 'Severe Depression'], x='No_of_hours_you_sleep', color='green', label='Severe Depression', kde=True,stat='density',weights=mild_depression_count/no_depression_count)

# Set labels and title
plt.xlabel('No of hours slept')
plt.ylabel('Frequency')
plt.title('Distribution of Hours Slept by Depression Category')
plt.legend(title='Depression Category')

# Show the plot
plt.show()
```

**1. Peak Sleep Duration**: The largest frequency for the 'No Depression' category (0) peaks at around 7 to 8 hours of sleep, which is commonly recommended as the healthy sleep duration for adults.

**2. Sleep Duration and Depression**: There's a noticeable pattern where individuals with 'No Depression' tend to have a more centralized and higher peak, suggesting they are more likely to have a consistent sleep duration of 7-8 hours.

**3. Disrupted Sleep in Depression**: Individuals in both the 'Mild' (1) and 'Severe' (2) depression categories exhibit more variability in sleep duration with multiple peaks, including significant frequencies at lower (3-5 hours) and higher (9-11 hours) sleep durations. This could indicate a link between depression and disrupted sleep patterns.

**4. Less Sleep and Severe Depression**: There are notable spikes for severe depression at the lower end of the sleep duration scale (around 3-5 hours). This might suggest that individuals with severe depression could be more likely to experience insomnia or other sleep disturbances.

**5. Excessive Sleep and Depression**: There are also peaks in the 'Severe Depression' category at higher sleep durations (9-10 hours). This could imply that hypersomnia, or excessive sleep, is also associated with more severe depression.

**6. Mild Depression Sleep Patterns**: Those with mild depression appear to have a wide spread of sleep durations with no single predominant peak, which might indicate variability in how mild depression affects sleep.

**7. Clinical Implications**: These patterns could be used by clinicians to better understand and inquire about sleep patterns in patients with depression, as sleep quantity may be an indicator of mental health status.

**8. Sleep Hygiene Interventions**: For public health, this graph supports the importance of sleep hygiene as a potential intervention area for mental health, particularly in preventing or treating depression.

In summary, the graph suggests a correlation between sleep duration and depression, with both very low and very high hours of sleep associated with higher frequencies of depression. This could inform both clinical practice and public health policies focusing on sleep as a factor in mental health.

**2. Low PIR and Depression**: There's a significant peak at the lower end of the PIR scale (close to 0) for severe depression (category '2'). This suggests a correlation where individuals closer to the poverty line experience higher rates of severe depression.

**Mid-range PIR**: The distribution for mild depression (category '1') is relatively spread out across the PIR range with no significant peaks, indicating that mild depression affects individuals across various income levels somewhat evenly.

**High PIR and Lower Depression Rates**: For both mild and severe depression categories, the frequency drops significantly as PIR increases, particularly noticeable in the PIR range above 4. This decline suggests that higher income, and thus a higher PIR, may be associated with lower reported rates of depression.

**Economic Stability and Mental Health**: The graph could imply that economic stability and being well above the poverty line might be protective factors against depression. This is in line with research that links financial stress to mental health issues.

**Implications for Social Policy**: The data might support the development of policies aimed at economic improvement as a means to potentially mitigate some mental health issues within the population. Focusing on poverty alleviation could be seen as part of a broader public health strategy.

**PIR as a Risk Factor**: PIR might be considered a risk factor to be monitored when assessing individual mental health, especially for identifying those at risk of severe depression.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Define the features representing symptoms
symptoms = ['Little_interest_in_doing_things', 'Feeling_down_or_depressed',
            'Trouble_sleeping_or_sleeping_too_much', 'Feeling_tired_or_having_little_energy',
            'Poor_appetite_or_overeating', 'Feeling_bad_about_yourself',
            'Trouble_concentrating_on_things', 'Moving_or_speaking_slowly_or_too_fast','Thought_you_would_be_better_off_dead']

# Create a new column representing the total count of symptoms for each patient
filtered_df['Total_Symptoms'] = filtered_df[symptoms].apply(lambda x: ((x == 1) | (x == 2) | (x == 3)).sum(), axis=1)

# Filter data for depressed patients
#depressed_data = df[df['depression_category'] > 0]

# Plotting
plt.figure(figsize=(10, 6))
sns.countplot(data=filtered_df, x='depression_category',hue='Total_Symptoms', palette='muted')

# Set labels and title
plt.xlabel('Number of Symptoms')
plt.ylabel('Count')
plt.title('Distribution of Symptom Counts Among Depressed Patients')
plt.xticks([0,1,2,3],labels=['No symptom','several days','More than half of days','Nearly everyday']))
# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plotting each feature against depression category
features = ['Little_interest_in_doing_things', 'Feeling_down_or_depressed',
            'Trouble_sleeping_or_sleeping_too_much', 'Feeling_tired_or_having_little_energy',
            'Poor_appetite_or_overeating', 'Feeling_bad_about_yourself',
            'Trouble_concentrating_on_things', 'Moving_or_speaking_slowly_or_too_fast','Thought_you_would_be_better_off_dead']

plt.figure(figsize=(15, 10))

for i, feature in enumerate(features, 1):
    plt.subplot(3, 3, i)
    sns.countplot(data=filtered_df, x=feature, hue='depression_category',palette='muted')
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.title(f'{feature} by Depression Category')
    plt.legend(title='Depression Category')

plt.tight_layout()
plt.show()
```
