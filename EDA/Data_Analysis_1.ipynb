##  **Prevalence of Sleep Apnea**


```python
no_apnea_count = data['Sleep_Apnea'].value_counts()
print(no_apnea_count)
max_count=max(no_apnea_count)
print(f'\nMax Count :{max_count}')
```

```python
normalized_counts = no_apnea_count / max_count

# Prevalence of Sleep Apnea
plt.figure(figsize=(5, 5))
sns.barplot(x=normalized_counts.index, y=normalized_counts.values, palette='coolwarm')
plt.title('Normalized Prevalence of Sleep Apnea')
plt.xticks(ticks=[0, 1], labels=['No', 'Yes'])
plt.xlabel('Sleep Apnea')
plt.ylabel('Density')
plt.show()
```

# **Depression vs Sleep Apnea**

```python
# Count of each category in "depression_category"
depression_count = data.groupby(['depression_category', 'Sleep_Apnea']).size().unstack(fill_value=0)

# Normalize the data by dividing each row by its sum
normalized_depression_count = depression_count.div(depression_count.sum(axis=1), axis=0)

# Plotting
normalized_depression_count.plot(kind='bar', stacked=False, figsize=(12, 4))


plt.title('Effects of Sleep Apnea in Depressed Patients (Normalized)')
plt.xlabel('\n Depression Category')
plt.ylabel('Proportion od Sleep apnea problems \n in each depressed Category')
plt.xticks(rotation=0)
plt.legend(title='Sleep Apnea', labels=['No Sleep Apnea', 'Sleep Apnea'])
plt.show()
```

#  **Depression vs Hypertension**

```python
#Count of each category in "depression_category"
depression_count = data.groupby(['depression_category', 'Blood_pressure']).size().unstack(fill_value=0)
depression_count = depression_count.reindex(['No Depression', 'Mild Depression', 'Severe Depression'])
print(depression_count)
# Normalize the data by dividing each row by its sum
normalized_depression_count = depression_count.div(depression_count.sum(axis=1), axis=0)

# Plotting
normalized_depression_count.plot(kind='bar', stacked=False, figsize=(14, 4))

plt.title('Effects of Blood Pressure in Depressed Patients (Normalized)')
plt.xlabel('\nDepression Category')
plt.ylabel('Proportion of Hypertension problems\n in each depressed Category')
plt.xticks(rotation=0)
plt.legend(title='Blood Pressure', labels=['No Hypertension', 'Hypertension'])
plt.show()
```

# **Depression vs Age**

```python
import pandas as pd

# Define the age bins
age_bins = range(20, 86, 10)  # Bins from 20 to 85 with a width of 5

# Create a new column 'Age_Bins' with the bin labels
data['Age_Bins'] = pd.cut(data['Age'], bins=age_bins, right=False)

# Count of each category in "depression_category" and "Age_Bins"
depression_count = data.groupby(['depression_category', 'Age_Bins']).size().unstack(fill_value=0)
depression_count = depression_count.reindex(['No Depression', 'Mild Depression', 'Severe Depression'])
print(depression_count)
# Normalize the data by dividing each row by its sum
normalized_depression_count = depression_count.div(depression_count.sum(axis=1), axis=0)

# Plotting
normalized_depression_count.plot(kind='bar', stacked=False, figsize=(20, 4))

plt.title('Effects of Age in Depressed Patients (Normalized)')
plt.xlabel('\nDepression Category')
plt.ylabel('Proportion of Age groups in each \n depressed Category')
plt.xticks(rotation=0)
plt.legend(title='Age Group')
plt.show()
```
