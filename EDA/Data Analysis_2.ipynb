# **Age Vs Depression**

```python
# Plot histograms for 'Age' with hue as 'depression_category'
sns.histplot(data=df[df['depression_category'] == 'No Depression'], x='Age', color='blue',  kde=True)
sns.histplot(data=df[df['depression_category'] == 'Mild Depression'], x='Age', color='yellow', kde=True)
sns.histplot(data=df[df['depression_category'] == "Severe Depression"], x='Age', color='green',  kde=True)

# Set labels and title
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Age by Depression Category')
plt.legend(title='Depression Category',labels=['No Depression','Mild Depression','Severe Depression'])

# Show the plot
plt.show()
```

**Normalised Plot**

```python
# Compute the count of 'No Depression' category to normalise histograms
no_depression_count = df[df['depression_category'] == 'No Depression']['Age'].count()
mild_depression_count = df[df['depression_category'] == 'Mild Depression']['Age'].count()
severe_depression_count = df[df['depression_category'] == 'Severe Depression']['Age'].count()


# Plot histograms for 'Age' with hue as 'depression_category'
sns.histplot(data=df[df['depression_category'] == 'No Depression'], x='Age', color='blue',  kde=True, stat='density',weights=no_depression_count/no_depression_count)
sns.histplot(data=df[df['depression_category'] == 'Mild Depression'], x='Age', color='yellow', kde=True,stat='density',weights=mild_depression_count/no_depression_count)
sns.histplot(data=df[df['depression_category'] == "Severe Depression"], x='Age', color='green',  kde=True,stat='density',weights=severe_depression_count/no_depression_count)

# Set labels and title
plt.xlabel('Age')
plt.ylabel('Density')
plt.title('Distribution of Age by Depression Category')
plt.legend(title='Depression Category',labels=['No Depression','Mild Depression','Severe Depression'])

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Depression-related symptoms
features = ['Little_interest_in_doing_things', 'Feeling_down_or_depressed',
            'Trouble_sleeping_or_sleeping_too_much', 'Feeling_tired_or_having_little_energy',
            'Poor_appetite_or_overeating', 'Feeling_bad_about_yourself',
            'Trouble_concentrating_on_things', 'Moving_or_speaking_slowly_or_too_fast',
            'Thought_you_would_be_better_off_dead']

rows = 3
cols = 3

# Create subplots
fig, axes = plt.subplots(rows, cols, figsize=(20, 15))

# Flatten axes for easy iteration
axes = axes.flatten()

for i, feature in enumerate(features):
    # Count of each category in "depression_category" and the current feature
    depression_count = df.groupby(['depression_category', feature]).size().unstack(fill_value=0)
    depression_count = depression_count.reindex(['No Depression', 'Mild Depression', 'Severe Depression'])

    # Normalize the data by dividing each row by its sum
    normalized_depression_count = depression_count.div(depression_count.sum(axis=1), axis=0)

    # Plotting
    normalized_depression_count.plot(kind='bar', stacked=False, ax=axes[i], legend=True)
    axes[i].set_xlabel(feature)
    axes[i].set_ylabel('Proportion')
    axes[i].set_title(f'{feature} by Depression Category')
    axes[i].legend(title='Depression Category')

# Adjust layout
plt.tight_layout()
plt.show()
```
